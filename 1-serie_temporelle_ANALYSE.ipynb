## Qu’est-ce qu’une série temporelle?


Les séries temporelles recouvrent un large éventail de phénomènes de la vie réelle et se retrouvent dans de nombreux domaines.

Une série temporelle, cela peut être aussi bien l’évolution de la population ou du PIB d’un pays, qu’un électrocardiogramme ou le dernier son de Dua Lipa.

Mathématiquement une série temporelle c’est une série de données indexée par le temps.

L’analyse et la prédiction de ces séries sont donc d’un intérêt primordial pour certaines industries ou secteurs d’activités car concrètement prédire une série temporelle c’est prédire le futur.

Une série temporelle est une séquence dans laquelle une métrique est enregistrée sur des intervalles de temps réguliers.

Selon la fréquence, une série temporelle peut être annuelle (ex : budget annuel), trimestrielle (ex : dépenses), mensuelle (ex : trafic aérien), hebdomadaire (ex : quantité de ventes), journalière (ex : météo), horaire (ex : cours des actions), minutes (ex : appels entrants dans un galop d'appel) et même secondes (ex : trafic Web).

## Comment se décompose une série temporelle ?


Traditionnellement une  série temporelle est décomposée selon trois éléments :

–        Une tendance (Tt)

–        Une saisonnalité (St)

–        Un résidu ou erreur (εt)

 Mathématiquement on peut donc traduire une série temporelle par  Xt = Tt  + St + εt

Avec  T la tendance, S la saisonnalité,  ε le résidu  et t l’indexation temporelle.

 La tendance

La tendance correspond à un comportement croissant ou décroissant d’une série au cours du temps.  Elle reflète souvent un phénomène de croissance ou décroissance sur le long terme.

La tendance d’une série temporelle peut prendre différentes formes :

–  Linéaire   Tt = α + βt

– Quadratique Tt = α + βt + γt²

-Exponentielle Tt = α + βexp(t)

….

La saisonnalité

La saisonnalité reflète la présence d’un phénomène périodique qui se répète au long de la série temporelle.

Ainsi, si la composante saisonnière se répète selon période k. On aura St+k = St

De nombreuse données présentent des saisonnalités, en particulier les données météorologiques (évolution de la température au cours du temps)

Certaines séries temporelles présentent à la fois une tendance et une saisonnalité, c’est le cas du trafic aérien mondial. On observe à la fois une croissance du trafic aérien tout en remarquant une différence forte entre le trafic hivernal et estival.


Source: DataScientest
Le résidu

Le résidu du modèle correspond à la partie de la série temporelle que la décomposition ne permet pas d’expliquer. On ne peut pas décomposer intégralement une série temporelle uniquement selon une tendance et une saisonnalité.

Idéalement le résidu du modèle est stationnaire, c’est-à-dire que le processus restant n’évolue pas avec le temps (moyenne et variance constante) . Si le résidu de notre série temporelle n’est pas stationnaire cela signifie que certaines composantes temporelles ne sont pas expliquées dans le modèle.

Une fois la tendance et la saisonnalité de la série temporelle expliquées, on peut donc chercher à expliquer le résidu de la décomposition avec des processus d’auto-régression ou de moyennes mobiles qui ont donné naissance au fameux modèle ARMA

Le graal de la modélisation des séries temporelles est d’obtenir un résidu de type bruit blanc, c’est-à-dire un résidu qui ne contient plus aucune information temporelle. En pratique c’est donc un signal stationnaire aléatoire et décorrélé.

Pour aller plus loin

– En plus de la saisonnalité, on définit quelquefois un cycle qui peut être considéré comme une saisonnalité de période plus longue, on peut alors définir plusieurs cycles différents.

 

– La présence d’une saisonnalité avérée entre deux dates peut rendre la comparaison entre ces dates difficiles. Selon les problématiques on peut donc chercher à corriger les variations saisonnières des séries temporelles. C’est le cas par exemple du taux de chômage ou du taux de croissance publiés par l’INSEE.

 

– La décomposition Xt = Tt  + St + εt est dite additive, on peut aussi modéliser une série temporelle selon une décomposition multiplicative  Xt = Tt (1 + St )(1 + εt )

D’où provient la non stationnarité ? Une décomposition très classique d’une série chronologique nous fournit quelques éléments de réponses.

Tendance: une tendance correspondant à une évolution de long terme de la série. Elle n’est pas forcément linéaire.

Saisonnalité : il s’agit de la propriété d’une série chronologique affichant des comportements périodiques se répétant à une fréquence constante (semaine/ week-end, été/hiver).

Cycles: Les cycles sont des saisons qui ne se produisent pas à fréquence fixe.

Bruit: Le bruit statistique correspond à des fluctuations de nature irrégulières, aléatoires et inexpliquées.

Cette décomposition peut être additive Y = T + S + C + epsilon ou 
multiplicative Y = T*S*C+epsilon. 
Il est également possible de combiner ces deux décompositions

## Conseils 1

Pour une compréhension approfondie de l'analyse statistique descriptive univariée de la consommation énergétique, voici le détail de chaque composante de l'analyse que vous pourriez faire :

1. **Moyenne** : La moyenne de la consommation énergétique, calculée comme la somme de toutes les consommations divisée par le nombre total d'observations, représente la consommation typique au sein de l'échantillon. C'est un indicateur central de tendance qui aide à comprendre le niveau général de consommation.

2. **Minimum et Maximum** : 
   - **Minimum** : La valeur la plus basse de consommation enregistrée, indiquant le niveau minimal de consommation d'énergie.
   - **Maximum** : La valeur la plus haute, reflétant le pic de consommation énergétique dans l'échantillon.

3. **Écart Type** : Cette mesure de dispersion calcule la variation moyenne des valeurs par rapport à la moyenne. Un écart type élevé indique une grande variabilité des consommations au sein de l'échantillon.

4. **Quartiles** :
   - **Premier Quartile (Q1)** : 25% des données sont inférieures ou égales à cette valeur, indiquant le seuil de consommation bas.
   - **Médiane (Q2)** : Aussi appelée le deuxième quartile, elle coupe l'ensemble de données en deux parties égales, montrant la valeur médiane de consommation.
   - **Troisième Quartile (Q3)** : 75% des données sont inférieures à cette valeur, marquant le seuil supérieur de consommation pour la majorité des observations.

5. **Données Manquantes** : Le nombre de valeurs manquantes peut affecter la précision des mesures statistiques. Il est important d'identifier ces lacunes pour comprendre les limitations des données.

6. **Données Dupliquées** : Les données qui apparaissent plus d'une fois peuvent indiquer des erreurs de saisie ou des duplications dans la collecte des données. Il est crucial de les identifier pour assurer l'intégrité de l'analyse.

7. **Valeurs Aberrantes** : Les valeurs extrêmes qui diffèrent significativement des autres données. Elles peuvent révéler des erreurs de mesure ou des cas exceptionnels qui pourraient fausser les analyses générales.

8. **Coefficient de Variation ou dispersion** : Il exprime l'écart type en pourcentage de la moyenne, fournissant un indicateur relatif de la variabilité des données, utile pour comparer des distributions avec des moyennes différentes.

9. **Test de Normalité (Shapiro-Wilk)** : Ce test statistique évalue si les données suivent une distribution normale. Une valeur-p inférieure à 0,05 suggère généralement que les données ne sont pas normalement distribuées.

10. **Stationnarité (Test de Dickey-Fuller augmenté)** : Ce test examine si une série temporelle est stationnaire, c'est-à-dire si ses propriétés statistiques ne varient pas dans le temps. Une valeur-p faible (typiquement <0,05) rejette l'hypothèse de non-stationnarité.

11. **Données Insignifiantes** : Observations qui ne changent pas ou peu et ne contribuent pas significativement à l'analyse ou aux insights, possiblement en raison de leur manque de variation.

12. **Problèmes de Format** : Erreurs dans la manière dont les données sont enregistrées ou présentées, nécessitant une standardisation pour assurer la comparabilité et la précision de l'analyse.

Chacune de ces composantes contribue à une compréhension plus nuancée et détaillée de la consommation énergétique, offrant des insights essentiels pour l'analyse et la prise de décisions.

### Bruit-hasard-aleatoire

Le terme de "bruit" dans un modèle statistique ou économétrique est essentiellement un terme générique pour décrire les variations ou perturbations aléatoires dans les données qui ne sont pas expliquées par le modèle utilisé. Voici quelques exemples concrets de ce que pourrait représenter ce bruit dans différents contextes :

1. **Météorologie** : Dans la prévision météorologique, le bruit pourrait représenter les fluctuations inattendues des conditions météorologiques qui ne sont pas capturées par les modèles standards. Cela pourrait inclure des variations imprévues des températures ou des précipitations dues à des phénomènes non modélisés.

2. **Finance** : Dans le domaine financier, le bruit peut refléter les mouvements quotidiens des prix des actions ou des autres instruments financiers qui ne sont pas expliqués par les modèles économiques classiques. Ces mouvements pourraient être dus à des nouvelles inattendues, des rumeurs, des changements dans le sentiment des investisseurs, ou d'autres facteurs exogènes.

3. **Économie** : En économie, le bruit peut comprendre des variations aléatoires dans les données économiques, comme le PIB ou le taux de chômage, qui ne sont pas prévus par les modèles économiques utilisés. Ces variations pourraient être causées par des événements non économiques, tels que des catastrophes naturelles ou des changements politiques soudains.

4. **Ingénierie** : Dans les applications d'ingénierie, telles que le contrôle des processus ou la surveillance de l'état des machines, le bruit peut être dû à des erreurs de mesure, des défauts d'équipement, ou des interférences externes qui affectent les lectures des capteurs.

5. **Recherche médicale** : En recherche médicale, le bruit dans les données expérimentales peut inclure des variations dans les réponses des patients à un traitement qui ne sont pas dues aux effets du traitement lui-même, mais plutôt à des différences individuelles, à des erreurs de dosage, ou à d'autres facteurs incontrôlables.

Dans tous ces exemples, le bruit est une partie inévitable des données collectées et analysées. Il est crucial pour les modélisateurs et les analystes de reconnaître l'existence de ce bruit et de tenter de le minimiser ou de le prendre en compte lors de l'interprétation des résultats des modèles.

## Granger Causality Test


```python
#faites attention avec les CORRELATIONS, dans le cas des series temporelles, utilisez les notions de CAUSALITE
#CAUSALITE: est ce que l'HISTORIQUE DE X permet de connaitre le FUTURE DE Y

#est ce que l'HISTORIQUE DE Y permet de connaitre le FUTURE DE Y  (la memoire, autoregression: LSTM)
```

```python
#correlation au sens de PEARSON Karl: est ce que X et Y sont liées? 
#              ou est ce que connaitre X permet de connaitre Y

#Causalité au sens de GRANGER: est ce l'historique de X permet de predire le future de Y
            #Vocabulaire: X cause Y au sens de Granger!
    
#Causalité au sens de NEWTON ou ARISTOTE: est ce que X impacte directemet Y
            #Vocabulaire: X impacte ou influe Y
```

```python
#h0:The Null hypothesis for grangercausalitytests is that the time series in the second column, X, does NOT Granger cause the time series in the first column, Y
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
#H0: il y'a pas de "causalité"   (X ne cause pas Y au sens de Granger )
#si pvalue<5% on rejette h0 et donc il y'a causalité
#
grangercausalitytests(data[["Consommation totale","Production totale"]],maxlag=12, verbose=False)
#est ce que la production CAUSE la consommation au cours des 12 derniers mois
#est ce que le fait de connaitre les 12 derniers productions nous donne des infos sur la consommation actuelle?
```

```python

maxlag=12   #conseil: prenez toujours la frequence de la serie temporelle
test = 'ssr_chi2test'

def grangers_causation_matrix(data, variables, test='ssr_chi2test', verbose=False):    
   
    df = pd.DataFrame(np.zeros((len(variables), len(variables))), columns=variables, index=variables)
    for c in df.columns:
        for r in df.index:
            test_result = grangercausalitytests(data[[r, c]], maxlag=maxlag, verbose=False)
            p_values = [round(test_result[i+1][0][test][1],4) for i in range(maxlag)]
            if verbose: print(f'Y = {r}, X = {c}, P Values = {p_values}')
            min_p_value = np.min(p_values)
            df.loc[r, c] = min_p_value
    df.columns = [var + '_x' for var in variables]
    df.index = [var + '_y' for var in variables]
    return df


```

## Cotintegration

Pour effectuer un **test de cointégration** entre la consommation et la production d'énergie en France, voici les étapes à suivre :

1. **Collecte des données** : Rassemblez des séries temporelles historiques sur la consommation et la production d'énergie en France. Des sources fiables comme le site du ministère de la Transition écologique fournissent ces données. citeturn0search2

2. **Tests de stationnarité** : Avant de procéder au test de cointégration, il est essentiel de vérifier si les séries sont stationnaires en utilisant des tests comme le test de Dickey-Fuller augmenté (ADF). Si les séries ne sont pas stationnaires en niveau mais le deviennent après différenciation, elles sont dites intégrées d'ordre 1, noté I(1).

3. **Test de cointégration** : Si les séries sont I(1), appliquez un test de cointégration, tel que le test de Johansen, pour déterminer s'il existe une relation d'équilibre à long terme entre la consommation et la production d'énergie.

4. **Interprétation des résultats** :
   - **Présence de cointégration** : Indique une relation stable à long terme entre la consommation et la production d'énergie. Cela signifie que, malgré les fluctuations à court terme, les deux séries évoluent conjointement sur le long terme.
   - **Absence de cointégration** : Suggère qu'il n'existe pas de relation d'équilibre à long terme entre ces variables, impliquant qu'elles évoluent indépendamment l'une de l'autre sur le long terme.

**Remarque** : L'analyse de cointégration est une méthode économétrique avancée qui nécessite une expertise spécifique. Il est recommandé de consulter un économètre ou un statisticien pour une analyse approfondie et précise. 

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.vector_ar.vecm import coint_johansen


# Sélection des séries de consommation et de production
consommation = data['Consommation totale']
production = data['Production totale']

# Test de stationnarité (ADF)
def test_adf(series, name):
    result = adfuller(series.dropna())
    print(f'Test ADF pour {name}:')
    print(f'Statistique ADF : {result[0]}')
    print(f'Valeur p : {result[1]}')
    print('---')

test_adf(consommation, 'Consommation')
test_adf(production, 'Production')

# Test de cointégration de Johansen
# Préparation des données pour le test
df_coint = data[['Consommation totale', 'Production totale']].dropna()

# Réalisation du test
johansen_test = coint_johansen(df_coint, det_order=0, k_ar_diff=1)

# Résultats du test
print('Valeurs propres :', johansen_test.eig)
print('Statistiques de trace :', johansen_test.lr1)
print('Valeurs critiques (90%, 95%, 99%) :', johansen_test.cvt)
```

### **Interprétation des résultats du test de stationnarité et de cointégration**

#### **1. Test ADF (Dickey-Fuller augmenté)**
Le **test ADF** permet de vérifier si une série temporelle est **stationnaire** ou non. L'hypothèse nulle (\(H_0\)) du test est que la série possède une **racine unitaire**, donc qu'elle est **non stationnaire**.

- **Consommation d'énergie** :  
  - **Statistique ADF** : -1.75  
  - **Valeur p** : **0.4068** (> 0.05)  

- **Production d'énergie** :  
  - **Statistique ADF** : -1.71  
  - **Valeur p** : **0.4244** (> 0.05)  

✅ **Interprétation** :  
- **Les deux séries ne sont pas stationnaires** car leurs **valeurs p sont supérieures à 0.05**.  
- Cela signifie que la consommation et la production d’énergie en France **présentent une tendance** et doivent être différenciées pour devenir stationnaires.  
- Il faut probablement **appliquer une différenciation première (différence entre valeurs consécutives)** et refaire le test ADF pour voir si les séries deviennent stationnaires.

---

#### **2. Test de cointégration de Johansen**
Le **test de cointégration** vérifie si deux (ou plusieurs) séries non stationnaires sont liées par une relation d'équilibre à long terme.

- **Valeurs propres** : **[0.2898, 0.2125]**  
- **Statistiques de trace** :  
  - Première statistique : **67.9954**  
  - Deuxième statistique : **27.9506**  

- **Valeurs critiques** à **95%** :  
  - Première statistique : **15.4943**  
  - Deuxième statistique : **3.8415**
